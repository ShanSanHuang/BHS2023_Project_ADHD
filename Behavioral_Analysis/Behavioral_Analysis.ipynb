```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

data = pd.read_csv('/Users/huangshansan/desktop/combine.re.csv')
data.keys()

# 1-back, Accuracy, Two-Way ANOVA
model_1back_ac = ols('Accuracy_1back~Diagnosed+Reward_timing+Diagnosed*Reward_timing', data=data).fit()
anova_1back_ac = sm.stats.anova_lm(model_1back_ac, typ=2)
anova_1back_ac
```

```python
# 2-back, Accuracy, Two-Way ANOVA
model_2back_ac = ols('Accuracy_2back~Diagnosed+Reward_timing+Diagnosed*Reward_timing', data=data).fit()
anova_2back_ac = sm.stats.anova_lm(model_2back_ac, typ=2)
anova_2back_ac
```

```python
# 1-back, Response_time, Two-Way ANOVA
model_1back_rt = ols('Response_time_1back~Diagnosed+Reward_timing+Diagnosed*Reward_timing', data=data).fit()
anova_1back_rt = sm.stats.anova_lm(model_1back_rt, typ=2)
anova_1back_rt
```

```python
# 2-back, Response_time, Two-Way ANOVA
model_2back_rt = ols('Response_time_2back~Diagnosed+Reward_timing+Diagnosed*Reward_timing', data=data).fit()
anova_2back_rt = sm.stats.anova_lm(model_2back_rt, typ=2)
anova_2back_rt
```
